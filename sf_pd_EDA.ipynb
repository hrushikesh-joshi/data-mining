### Data Cleaning

#### Dealing with NaN data

Removing columns with more than 70% missing values and dropping CAD Number and geo location columns and then removing NaNs.

```python
#dropping columns that are not significant for future data exploration
df.drop(['Invest In Neighborhoods (IIN) Areas','ESNCAG - Boundary File','Civic Center Harm Reduction Project Boundary'],axis='columns', inplace=True)

df.drop(['Central Market/Tenderloin Boundary Polygon - Updated','Filed Online','HSOC Zones as of 2018-06-05','CAD Number','Latitude'], axis='columns', inplace=True)

df.drop(['Longitude','Point'], axis='columns', inplace=True)

df.dropna(inplace=True)

print('Clean dataset: ')
print("Observations: ", df.shape[0])
print("Variables: ", df.shape[1])

#examing the changes
df.head(5)
```

 #### Preprocessing Text about Crime Type
 
 - The `unique()` function can extract a list of distinct values in a dataset, it returns the sorted unique elements of an array.
 
unique values of police districts


```python
len(df['Police District'].unique())
```

- create a new column to calculate how long does it take for a crime to be reported

```python
from datetime import timedelta
df['Report Datetime'] = pd.to_datetime(df['Report Datetime'])
df['Incident Datetime'] = pd.to_datetime(df['Incident Datetime'])
df['Response Time'] = (df['Report Datetime'] - df['Incident Datetime']) / timedelta(minutes=1)
```

```python
df['Response Time']
```

```python
from matplotlib import pyplot as plt
df.groupby('Police District')['Response Time'].mean().plot(kind='bar')
plt.title('Average response time by Police District')
plt.xlabel('Police District')
plt.ylabel('Response time (minutes)')
plt.show()
```

We can see that response to crimes out side the SF district is higher which is understandable. But there is still room for improvement is some police districts such as Northern and Taraval.

```python
from matplotlib import pyplot as plt
df.groupby('Incident Category')['Response Time'].mean().plot(kind='bar')
plt.title('Average response time by Incident Category')
plt.xlabel('Incident Category')
plt.ylabel('Response time (minutes)')
plt.show()
```

We can see that human trafficing(B) involuntary servitude has a response time that goes years and there is a severe need to address this crime as victims go without justice for a long period of time.

## Exploratory Analysis
### 1. Crimes by year and day


```python
df['Incident Year'].value_counts().plot(kind="bar", title = "Total Crime Events by Year")
```

We can observe that crime was on the decline from 2018 to 2020 but there has been a rise in crime from 2021 onwards

```python
df['Incident Day of Week'].value_counts().plot(kind="bar", title = "Total Crime Events by Year")
```
